```python
import os, random
train_img_total, test_img_total, val_img_total = [], [], []
train_lbl_total, test_lbl_total, val_lbl_total = [], [], []

############## 수정 할 부분: path, path2, random.sample(개수 몇개 추출할건지)


# 이미지 경로 입력(path)
path = "/content/drive/MyDrive/export/images/"
img_list = os.listdir(path)
img_list = [file for file in img_list if file.endswith(".jpg")]

# 클래스마다 이미지 분류
tan_0 = [x for x in img_list if "a7" in x or "a11" in x]
white_1 = [x for x in img_list if "a2" in x or "a4" in x or "a5" in x or "a8" in x or "a9" in x]
non_2 = [x for x in img_list if "b6" in x]
nop_3 = [x for x in img_list if "b7" in x]
nok_4 = [x for x in img_list if "b8" in x]
cal_5 = [x for x in img_list if "b3" in x]
gray_6 = [x for x in img_list if "a1" in x or "a6" in x]
nogyun_8 = [x for x in img_list if "a3" in x or "a12" in x]

# 분류한 클래스에서 랜덤으로 몇개 꺼내서 cnt 변수에 저장
#### 이거 수정!!!!!!!!! 
# tan_0_cnt = random.sample(tan_0,5)
# white_1_cnt = random.sample(white_1,5)
# non_2_cnt = random.sample(non_2,5)
# nop_3_cnt = random.sample(nop_3,30)
# nok_4_cnt = random.sample(nok_4,30)
# cal_5_cnt = random.sample(cal_5,30)
# gray_6_cnt = random.sample(gray_6,30)
# nogyun_8_cnt = random.sample(nogyun_8,30)

# img_list_total에 전부 저장
img_list_total = [tan_0, white_1, non_2, nop_3, nok_4, cal_5, gray_6, nogyun_8]
# img_list_total = [tan_0_cnt, white_1_cnt, non_2_cnt, nop_3_cnt, nok_4_cnt, cal_5_cnt, gray_6_cnt, nogyun_8_cnt]

# 클래스마다 랜덤으로 train(0.8), test(0.1), val(0.1) 저장 
for all_list in img_list_total:
    lens = len(all_list)

    if lens != 0:
        train_img = random.sample(all_list, int(lens * 0.8))
        train_img_total.extend(train_img)

# train_img에 들어간 이미지 빼고 랜덤 추출
        test_img = random.sample([x for x in all_list if x not in train_img_total], int(lens * 0.1))  
        test_img_total.extend(test_img)

# train_img, test_img에 들어간 이미지 빼고 랜덤 추출
        val_img = random.sample([x for x in all_list if x not in train_img_total and x not in test_img_total],int(lens * 0.1))
        val_img_total.extend(val_img)

# 정렬
train_img_total.sort()
test_img_total.sort()
val_img_total.sort()


# 텍스트 경로 입력(path2)
path2 = "/content/drive/MyDrive/export/labels/"
lbl_list = os.listdir(path2)
lbl_list = [file for file in lbl_list if file.endswith(".txt")]

# 정렬
lbl_list.sort()

# train, test, val 이미지가 저장된 이름과 같은 텍스트파일 찾아서 각각 저장
for lb in lbl_list:
    for h in train_img_total:
        if h[:-4] == lb[:-4]:
            train_lbl_total.append(lb)
    for o in test_img_total:
        if o[:-4] == lb[:-4]:
            test_lbl_total.append(lb)
    for p in val_img_total:
        if p[:-4] == lb[:-4]:
            val_lbl_total.append(lb)



# 정렬(확인을 위함)
# train_lbl_total.sort()
# test_lbl_total.sort()
# val_lbl_total.sort()

# print("1111111111")
# print(len(train_img_total))
# print(len(train_lbl_total))
# print(len(test_img_total))
# print(len(test_lbl_total))
# print(len(val_img_total))
# print(len(val_lbl_total))

print("222222222222")
print(train_img_total[:5])
print(train_lbl_total[:5])
print(test_img_total[:5])
print(test_lbl_total[:5])
print(val_lbl_total[:5])
print(val_img_total[:5])

# print("3333333333333")
# print(train_img_total[-5:])
# print(train_lbl_total[-5:])
# print(test_img_total[-5:])
# print(test_lbl_total[-5:])
# print(val_lbl_total[-5:])
# print(val_img_total[-5:])
```
